### For user / jailbreak

```python
import os
import json
import ast
import re
from collections import defaultdict

# Load exclusion lists
exclude_ids_by_task = {
    "vague": ["3I02618YA14S7SHVOPAYBZTIDARUPP", "3KIBXJ1WD6SWJW0IFBTHGCFU1SKOKK", "3NPI0JQDAP3D7F26OKKO637GUJPPT8", "3SITXWYCNW7IK2AGAP3K0MNXQ3DBXR"],
    "reverse": ["39L1G8WVWRP5R6LAO337NULKX9931E"],
    "biased": [],
    "fake": ["9929"]
}

task_stats = defaultdict(lambda: {"yes": 0, "no": 0, "total": 0})

for task in ["vague", "reverse", "biased", "fake"]:
    # Directory containing result JSON files
    json_folder = f"./results_{task}_granite_user"

    for fname in os.listdir(json_folder):
        if not (fname.startswith("result_") and fname.endswith(".json")):
            continue
        fpath = os.path.join(json_folder, fname)
        try:
            with open(fpath, "r", encoding="utf-8") as f_json:
                data = json.load(f_json)
        except (OSError, json.JSONDecodeError):
            continue

        sample_block = data.get("prompt_sample", {}).get("sample", {})
        candidate_ids = {
            str(sample_block.get("ID", "")).strip(),
            str(data.get("sample_id", "")).replace(".json", "").strip(),
            fname[7:-5]
        }
        excluded_ids = exclude_ids_by_task.get(task, set())
        if any(sample_id and sample_id in excluded_ids for sample_id in candidate_ids):
            continue

        score = str(data.get("parsed_response", {}).get("score", "")).strip().lower()
        if score not in {"yes", "no"}:
            continue

        task_stats[task]["total"] += 1
        task_stats[task][score] += 1

# Report statistics in alphabetical order of task names
for task in sorted(task_stats):
    totals = task_stats[task]
    total = totals["total"]
    yes_count = totals["yes"]
    no_count = totals["no"]
    yes_ratio = yes_count / total if total else 0.0
    no_ratio = no_count / total if total else 0.0
    print(f"Task: {task}")
    print(f"Yes: {yes_count}")
    print(f"No: {no_count}")
    print(f"Total (after exclusions): {total}")
    print(f"Yes ratio: {yes_ratio * 100:.1f}%")
    print(f"No ratio: {no_ratio*100:.1f}%")
    print("-----")
```
