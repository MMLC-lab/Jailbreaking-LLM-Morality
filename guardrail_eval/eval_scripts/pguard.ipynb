```python
import json
import os
from collections import defaultdict
import pandas as pd

def analyze_folder(folder_path):
    """Analyze predicted_label values across all JSON files in the given folder."""
    results = {"safe": 0, "jailbreak": 0, "other": 0, "missing": 0}
    file_count = 0
    errors = []
    
    if not os.path.exists(folder_path):
        return results, 0, [f"Folder not found: {folder_path}"]
    
    # Get all JSON files
    json_files = [f for f in os.listdir(folder_path) if f.endswith('.json')]
    
    for filename in json_files:
        file_count += 1
        filepath = os.path.join(folder_path, filename)
        
        try:
            with open(filepath, 'r', encoding='utf-8') as f:
                data = json.load(f)
            
            # Check for parsed_response.predicted_label
            if 'parsed_response' in data and 'predicted_label' in data['parsed_response']:
                label = data['parsed_response']['predicted_label'].lower().strip()
                if label == 'safe':
                    results['safe'] += 1
                elif label == 'jailbreak':
                    results['jailbreak'] += 1
                else:
                    # print(f"Unknown label '{label}' in file {filename}")
                    results['other'] += 1
            else:
                results['missing'] += 1
                
        except Exception as e:
            errors.append(f"Error processing file {filename}: {str(e)}")
            
    return results, file_count, errors

print("Starting analysis of JSON files across the four folders...")
```

```python
# Define the four folder paths
folders = {
    'results_vague_pguard': 'results_vague_pguard',
    'results_biased_pguard': 'results_biased_pguard',
    'results_fake_pguard': 'results_fake_pguard',
    'results_reverse_pguard': 'results_reverse_pguard'
}

# Store all results
all_results = {}

# Analyze each folder
for folder_name, folder_path in folders.items():
    print(f"\nAnalyzing folder: {folder_name}")
    print("=" * 50)
    
    results, file_count, errors = analyze_folder(folder_path)
    all_results[folder_name] = {
        'results': results,
        'file_count': file_count,
        'errors': errors
    }
    
    print(f"Total files: {file_count}")
    print(f"Safe labels: {results['safe']}")
    print(f"Unsafe labels: {results['jailbreak']}")
    print(f"Other labels: {results['other']}")
    print(f"Missing labels: {results['missing']}")
    
    # calculate percentages
    total_valid = results['safe'] + results['jailbreak'] + results['other']
    if total_valid > 0:
        safe_pct = (results['safe'] / total_valid) * 100
        unsafe_pct = (results['jailbreak'] / total_valid) * 100
        print(f"Safe percentage: {safe_pct:.1f}%")
        print(f"Unsafe percentage: {unsafe_pct:.1f}%")
```
